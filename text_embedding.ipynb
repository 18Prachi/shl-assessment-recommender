```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
import pickle

# Load enriched CSV
df = pd.read_csv("shl_enriched_catalog.csv")

# Fill NaNs in description
df["description"] = df["description"].fillna("")

# Load sentence-transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for descriptions
embeddings = model.encode(df["description"].tolist(), show_progress_bar=False)

# Save embeddings and catalog
np.save("shl_embeddings.npy", embeddings)
df.to_csv("shl_final_catalog.csv", index=False)
print("✅ Saved: shl_embeddings.npy and shl_final_catalog.csv")

# (Optional) Save catalog + embeddings together for easy loading later
with open("shl_catalog_with_embeddings.pkl", "wb") as f:
    pickle.dump((df, embeddings), f)
```
